## Задание 1
Скачайте с сайта grouplens.org...movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
```

```python
path = 'C:\\Users\\Сергей\\Downloads\\'
rating= pd.read_table(path+"ratings.dat",sep='::',
                   names=['UserID','MovieID','Rating','Timestamp'],engine='python')
names = pd.read_table(path+"movies.dat",sep='::', names=['MovieID','Title','Genres'],engine='python')
```

```python
top_movie = (rating.query('Rating ==5').groupby(['MovieID'])['Rating']
            .count().sort_values(ascending = False).reset_index()['MovieID'][0])
print('Больше всего оценок 5.0 было выставленно фильму -',names.query('MovieID ==@top_movie')['Title'].values[0])
```

## Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
power = pd.read_csv(path+"power.csv")
```

```python
countries = ['Latvia','Lithuania','Estonia']
categories = [4, 12, 21]
print('Cуммарное потребление стран Прибалтики - ',power.query(
    'country in @countries & category in @categories & 2005 <=year <=2010')['quantity'].sum())
```
