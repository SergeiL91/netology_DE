```python
import json
```

```python
keys = []
values = []
with open('purchase_log.txt', encoding="utf-8") as f:
    for line in f:
        line = line.strip()        
        dict_ = json.loads(line)
        for index, i  in enumerate(dict_.values()):
            if index == 0:
                keys.append(i)
            else:
                values.append(i)
    dict_ =  dict(zip(keys[1:], values[1:]))
```

```python
with open('results.txt', 'w') as result:
    with open('visit_log.csv', encoding="utf-8") as f:
        for line in f:
            line = line.strip().split(',') 
            for key,value in dict_.items():
                if key == line[0]:
                    print(','.join([str(elem) for elem in line]) +','+ value)
                    
```

```python
with open('funnel.csv', 'w') as funnel:
    with open('visit_log.csv', encoding="utf-8") as f:
        for line in f:
            line = line.strip().split(',') 
            for key,value in dict_.items():
                if key == line[0]:
                    final = ','.join([str(elem) for elem in line]) +','+ value
                    funnel.write(final)
```
